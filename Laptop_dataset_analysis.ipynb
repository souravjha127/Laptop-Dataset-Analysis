```python
#to remove or drop the unnamed column
df=df.drop(columns = 'Unnamed: 0' )
```

```python
df
```

```python
#removing the gb from the ram colmn and keeping only integer value
df["Ram"]=df["Ram"].str.replace('GB',' ')
```

```python
df
```

```python
#coverting the data of the Ram column from string to int
df['Ram'].astype('int64')
df.info()
```

```python
df['Weight'] = df['Weight'].str.replace('kg','')
df
```

```python
df['Weight']=df['Weight'].astype('float64')
df.info()
```
